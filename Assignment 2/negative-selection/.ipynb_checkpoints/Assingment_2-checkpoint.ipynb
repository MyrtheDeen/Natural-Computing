```python
import subprocess
import numpy as np
import matplotlib.pyplot as plt

def run_java_command(command):
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    print("Command:", command)
    print("Output:", result.stdout.strip())
    print("Error:", result.stderr.strip())
    return result.stdout.strip()

def calculate_auc(predictions):
    # Assuming predictions is a list of numerical values
    sorted_predictions = sorted(predictions, reverse=True)
    auc = np.trapz(sorted_predictions)
    return auc

#initialize folder
initialize_command = "cd '/Users/ringo/Library/CloudStorage/GoogleDrive-jasperpieterse12@gmail.com/Other computers/Desktop/Desktop/Physicss/Natural Computing/Assignments/Assignment 2/negative-selection' "
run_java_command(initialize_command)

# Train the data
training_command = "java -jar negsel2.jar -alphabet file://english.train -self english.train -n 10 -r 4 -c -l"
run_java_command(training_command)

# Run Java commands and capture output
output_english_test = run_java_command("java -jar negsel2.jar -self english.train -n 10 -r 4 -c -l < english.test")
output_tagalog_test = run_java_command("java -jar negsel2.jar -self english.train -n 10 -r 4 -c -l < tagalog.test")

# Process output into numerical values
english_test_values = list(map(float, output_english_test.split()))
tagalog_test_values = list(map(float, output_tagalog_test.split()))

# Calculate AUC
auc_english_test = calculate_auc(english_test_values)
auc_tagalog_test = calculate_auc(tagalog_test_values)

# Print or use AUC values as needed
print("AUC for English Test:", auc_english_test)
print("AUC for Tagalog Test:", auc_tagalog_test)
```
